```python
data = pd.read_json('combined.json', lines=True)
data = pd.DataFrame(data)
data
```

```python
data = data[data["topics"].str.len() != 0]
data.head(5)
```

```python
def preprocess(text):
    doc1 = nlp(text)
    no_stop_words = [token1.text for token1 in doc1 if not token1.is_stop and not token1.is_punct]
    return ' '.join(no_stop_words)
```

```python
miniData = data[0:100]
miniData
```

```python
miniData["process_content"]=miniData["contents"].apply(preprocess)
miniData.head(5)
```

```python
testData = miniData["process_content"]
testData
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v=TfidfVectorizer()
data_out =v.fit_transform(testData)
data_out
```

```python
v.vocabulary_
```

```python
all_feature = v.get_feature_names_out()
for feature in all_feature:
    index = v.vocabulary_.get(feature)
    print(f"{feature} {v.idf_[index]}")
```
