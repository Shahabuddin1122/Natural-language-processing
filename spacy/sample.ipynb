```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')

# Process the sentence
doc = nlp(sentence)

for token in doc:
    print(token.text, " | ",token.pos_," | ", token.lemma_)
```

```python
processData=[]
for token in doc:
    if token.pos_ not in ['SPACE','X', 'PUNCT','DET','AUX']:
        processData.append(token.text)

print(processData)
```

```python
word = [ token for token in processData ]
word
```

```python
import nltk
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
word = [stemmer.stem(word) for word in word]
word
```
